#  Conclusiones e Insights

A partir del análisis exploratorio de los datos de TelecomX se identificaron varios factores que pueden estar relacionados con la cancelación del servicio por parte de los clientes (Churn).

## 1. Distribución del abandono
El análisis inicial muestra que aproximadamente el 26.6% de los clientes han cancelado el servicio, mientras que el 73.4% continúan utilizando la plataforma. Esto indica que, aunque la mayoría de los clientes permanece, existe un porcentaje considerable de abandono que puede representar pérdidas importantes para la empresa.

## 2. Relación entre cargos mensuales y abandono
El análisis de los cargos mensuales muestra que los clientes que cancelan el servicio tienden a tener cargos mensuales más altos en comparación con aquellos que permanecen en la empresa. Esto sugiere que el precio del servicio puede ser un factor relevante en la decisión de cancelar.

## 3. Influencia del tiempo de permanencia (Tenure)
Se observó que los clientes con mayor tiempo en la empresa presentan una menor probabilidad de cancelar el servicio. Por el contrario, los clientes más recientes son quienes presentan mayores tasas de abandono. Esto indica que la fidelización del cliente en los primeros meses es un factor clave para reducir el churn.

## 4. Evolución del precio con el tiempo
Al analizar los cargos mensuales en función del tiempo de permanencia se identificó que los precios pueden aumentar ligeramente con el tiempo, aunque este incremento no es excesivo. Sin embargo, los clientes con mayores cargos mensuales siguen mostrando mayor tendencia al abandono.

## 5. Impacto del número de servicios contratados
Se encontró una correlación negativa entre el número de servicios contratados y el churn. Esto significa que los clientes que utilizan más servicios dentro de la empresa tienen menor probabilidad de cancelar su suscripción.

#  Recomendaciones

A partir de los resultados obtenidos se pueden plantear algunas estrategias para reducir el churn:

- Implementar estrategias de fidelización durante los primeros meses de servicio.
- Analizar la estructura de precios para evitar que los cargos mensuales elevados generen cancelaciones.
- Incentivar la contratación de múltiples servicios dentro de la plataforma.
- Desarrollar programas de retención para clientes con alto riesgo de abandono.

Estas estrategias podrían ayudar a mejorar la retención de clientes y reducir el impacto del churn en la empresa.
